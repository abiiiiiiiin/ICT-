# **Diabetes dataset EDA**

````python
db=pd.read_csv('/content/diabetes_model.csv')
db
````

# univariate analysis

````python
import seaborn as sns
db.hist(bins=15, figsize=(15, 10), layout=(3, 3))
plt.suptitle('Histograms of Diabetes Dataset Variables')
plt.show()
````

````python
db.plot.box(subplots=True, figsize=(15, 10),layout=(3, 3))
plt.suptitle('Box Plots of Diabetes Dataset Variables')
plt.show()
````

# **Bivariate Analysis**

````python
correlation_matrix = db.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
````

````python
sns.pairplot(db)
plt.show()
````
